```python
import pandas as pddf = pd.read_csv(r"C:\Users\kvred\OneDrive\Desktop\PROJECT DATA\helpdesk.csv")df['CompleteTimestamp'] = pd.to_datetime(df['CompleteTimestamp'])df.head()
```

```python
df = df.sort_values(['CaseID', 'CompleteTimestamp'])case_start = df.groupby('CaseID')['CompleteTimestamp'].min()case_end = df.groupby('CaseID')['CompleteTimestamp'].max()cases = pd.DataFrame({    'CaseID': case_start.index,    'start': case_start.values,    'end': case_end.values})cases['resolution_hours'] = (cases['end'] - cases['start']).dt.total_seconds() / 3600cases['sla_met'] = cases['resolution_hours'] <= 48
```

```python
cases = pd.DataFrame({    'CaseID': case_start.index,    'start': case_start.values,    'end': case_end.values})cases['resolution_hours'] = (cases['end'] - cases['start']).dt.total_seconds() / 3600cases['sla_met'] = cases['resolution_hours'] <= 48
```

```python
avg_resolution = cases['resolution_hours'].mean()median_resolution = cases['resolution_hours'].median()sla_rate = cases['sla_met'].mean() * 100total_cases = cases.shape[0]print(f"Total cases: {total_cases}")print(f"Average resolution (hrs): {avg_resolution:.2f}")print(f"Median resolution (hrs): {median_resolution:.2f}")print(f"SLA compliance (% <= 48 hrs): {sla_rate:.1f}%")
```

```python
import matplotlib.pyplot as pltimport seaborn as snssns.set(style="whitegrid")import osos.makedirs("visuals", exist_ok=True)# 1. Resolution time distributionplt.figure(figsize=(10,5))sns.histplot(cases['resolution_hours'], bins=40, kde=True, color="#4C78A8")plt.title("Case Resolution Time Distribution")plt.xlabel("Hours"); plt.ylabel("Case count")plt.tight_layout()plt.savefig("visuals/resolution_time_distribution.png")plt.show()# 2. SLA compliance pie chartplt.figure(figsize=(6,6))cases['sla_met'].value_counts().rename(index={True:"Met SLA", False:"Breached SLA"}).plot.pie(    autopct='%1.1f%%', colors=["#59A14F", "#E15759"])plt.title("SLA Compliance"); plt.ylabel("")plt.tight_layout()plt.savefig("visuals/sla_compliance.png")plt.show()
```
